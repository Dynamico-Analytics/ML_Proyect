## Tratamiento de valores nulos

### 1era Opción: Eliminar todos los valores nulos

```python
# Eliminar filas que contienen al menos un valor nulo
df_sin_nulos = salaries.dropna()

# Eliminar columnas que contienen al menos un valor nulo
df_sin_nulos_columnas = salaries.dropna(axis=1)
```

## Transformacón de datos

### 2da Opción: Reemplazar NaNs por la media

### Parsing ("Parseo") de la columna Salary

```python
df_sin_nulos['salary']
```

```python
## Este código toma una columna ´'salary'´ que contiene valores de salario en formato de cadena,
## elimina cualquier careacter no numérico, extrae los dígitos y convierte la columna resultante en valores numéricos de tipo float
## El resultado se almacena en una nueva columna llamada ´'salary_numeric´' 
## Esto es útil cuando la columna ´'salary´' inicial contiene caracteres no numéricos y se desea convertir a valores numéricos.

df_sin_nulos['salary_numeric'] = df_sin_nulos['salary'].replace('[^\\d.]', '', regex=True).str.extract('(\\d+)').astype(float)
```

```python
df_sin_nulos['salary_numeric']
```

```python
df_sin_nulos['salary_numeric'].describe()
```

```python
# Crear un boxplot para visualizar la distribución de los salarios
plt.boxplot(df_sin_nulos['salary_numeric'])

# Añadir etiquetas y titulo
plt.xlabel('Salario')
plt.ylabel('Valores en USD')
plt.title('Boxplot')

# Mostrar el boxplot
plt.show()
```

```python
# Crear un histograma para visualizar la distribución de los salarios
plt.hist(df_sin_nulos['salary_numeric'], bins=30, color='blue', alpha=0.7)

# Añadir etiquetas y titulo
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Histograma de Datos Aleatorios')

# Mostrar el histograma
plt.show()
```

```python
# El problema que vemos es que el dataset toma como referencia sueldos anuales, mensuales y por hora, por lo que es necesario normalizar los datos
# Eliminamos las filas que contengan un salario mayor o igual a 20000
df_sin_nulos = df_sin_nulos[df_sin_nulos['salary_numeric'] >= 20000]
```

```python
# Crear un boxplot para visualizar la distribución de los salarios
plt.boxplot(df_sin_nulos['salary_numeric'])

# Añadir etiquetas y titulo
plt.xlabel('Salario')
plt.ylabel('Valores en USD')
plt.title('Boxplot')

# Mostrar el boxplot
plt.show()
```

```python
# Crear un histograma para visualizar la distribución de los salarios
plt.hist(df_sin_nulos['salary_numeric'], bins=30, color='blue', alpha=0.7)

# Añadir etiquetas y titulo
plt.xlabel('Valores')
plt.ylabel('Frecuencia')
plt.title('Histograma de Datos Aleatorios')

# Mostrar el histograma
plt.show()
```

### Método de reemplazo de valores nulos con la media de la columna

```python
# Rellenar valores nulos con un valor específico (por ejemplo: el promedio)
df_con_relleno = df_sin_nulos.fillna(df_sin_nulos['salary_numeric'].mean())
```

### Eliminar la columna salary deprecada

```python
df_con_relleno.drop('salary', axis=1, inplace=True)
```

```python
df_con_relleno
```

### Agrupar y agregar datos

```python
# Supongamos que tienes un DataFrame llamado df con columnas 'title' y 'salary'
# Agrupar por título y calcular el promedio y la mediana del salario para cada titulo
df_grouped = df_con_relleno.groupby('title')['salary_numeric'].agg(['mean', 'median', 'count']).reset_index()

# Renombrar las columnas para mayor claridad
df_grouped.columns = ['title', 'salary_mean', 'salary_median', 'job_count']

# Mostrar el DataFrame resultante
df_grouped
```

```python
df_grouped['salary_mean'] = df_grouped['salary_mean'].round(0).astype(int)
df_grouped['salary_median'] = df_grouped['salary_median'].round(0).astype(int)
```

```python
df_grouped
```

### Exportar CSV

```python
# Exportar CSV
# Especifica la ruta y el nombre del archivo CSV
ruta_y_nombre_archivo_csv = "export-linkedin-jobs-usa-actualizado.csv"

# Exporta el DataFrame a un archivo CSV
df_grouped.to_csv(ruta_y_nombre_archivo_csv)
```
